THEORY QUESTIONS

1. What is a random variable in probability theory?
- A random variable assigns numerical values to the outcomes of a random experiment.

2. What are the types of random variables?
- Discrete and continuous.

3. What is the difference between discrete and continuous distributions?
- Discrete distributions deal with countable outcomes; continuous deal with uncountable, real-valued outcomes.

4. What are probability distribution functions (PDF)?
- A function that describes the likelihood of a random variable taking specific values.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
-CDF gives the probability a variable is less than or equal to a value; PDF gives the probability density at that value.

6. What is a discrete uniform distribution?
- A distribution where all outcomes have equal probability.

7. What are the key properties of a Bernoulli distribution?
- Only two outcomes (success/failure), with a single trial and a fixed probability of success.

8. What is the binomial distribution, and how is it used in probability?
- It models the number of successes in a fixed number of independent Bernoulli trials.

9. What is the Poisson distribution and where is it applied?
- Models the number of events in a fixed interval of time/space with a known constant rate.

10. What is a continuous uniform distribution?
- A distribution where all values in an interval are equally likely.

11. What are the characteristics of a normal distribution?
- Symmetrical, bell-shaped curve defined by mean and standard deviation.

12. What is the standard normal distribution, and why is it important?
- A normal distribution with mean 0 and SD 1; used to standardize values (Z-scores).

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- It states that the sampling distribution of the sample mean approaches normality as sample size increases, regardless of the original distribution.

14. How does the Central Limit Theorem relate to the normal distribution?
- It explains why normal distribution often appears in sampling, even if the data is not normal.

15. What is the application of Z statistics in hypothesis testing?
- Z-statistics determine how far a sample statistic is from the population mean under the null hypothesis.

17. What are point estimates and interval estimates in statistics?
- Point estimate gives a single value; interval estimate gives a range within which the parameter lies.

18. What is the significance of confidence intervals in statistical analysis?
- They indicate the range of values likely to contain the population parameter.

19. What is the relationship between a Z-score and a confidence interval?
- Z-scores define the boundaries of confidence intervals in a normal distribution.

20. How are Z-scores used to compare different distributions?
- By standardizing values, Z-scores allow comparison across distributions with different scales.

21. What are the assumptions for applying the Central Limit Theorem?
- Large sample size, independent samples, and identically distributed variables with finite mean and variance.

22. What is the concept of expected value in a probability distribution?
- The average or mean value a random variable takes in the long run.

23. How does a probability distribution relate to the expected outcome of a random variable?
- The expected outcome is computed using the probabilities and values defined in the distribution.

```python
#1. Write a Python program to generate a random variable and display its value.
import random

random_variable = random.randint(1, 100)
print("Random Variable:", random_variable)
```

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randint(1, 7, 10000)
values, counts = np.unique(data, return_counts=True)

pmf = counts / sum(counts)

plt.bar(values, pmf)
plt.title("PMF of Discrete Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.show()
```

```python
#Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    return p if x == 1 else 1 - p

print(bernoulli_pdf(1, 0.6))  # Example: P(X=1)
```

```python
#Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

data = np.random.binomial(n=10, p=0.5, size=10000)
plt.hist(data, bins=11, density=True, edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
#Create a Poisson distribution and visualize it using Python.
from scipy.stats import poisson

data = poisson.rvs(mu=3, size=10000)
plt.hist(data, bins=15, density=True, edgecolor='black')
plt.title("Poisson Distribution (Œª=3)")
plt.xlabel("Events")
plt.ylabel("Probability")
plt.show()
```

```python
#Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
data = np.sort(np.random.randint(1, 7, 1000))
cdf = np.arange(1, len(data)+1) / len(data)

plt.plot(data, cdf, marker='.', linestyle='none')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("CDF")
plt.grid(True)
plt.show()
```

```python
#Generate a continuous uniform distribution using NumPy and visualize it.
data = np.random.uniform(0, 1, 10000)
plt.hist(data, bins=50, density=True, edgecolor='black')
plt.title("Continuous Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
#Simulate data from a normal distribution and plot its histogram.
data = np.random.normal(loc=0, scale=1, size=10000)
plt.hist(data, bins=50, density=True, edgecolor='black')
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
#Write a Python function to calculate Z-scores from a dataset and plot them.
from scipy.stats import zscore

data = np.random.normal(50, 10, 100)
z_scores = zscore(data)

plt.plot(z_scores, marker='o')
plt.title("Z-scores of Dataset")
plt.xlabel("Index")
plt.ylabel("Z-score")
plt.grid(True)
plt.show()
```

```python
#Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
samples = [np.mean(np.random.exponential(scale=2.0, size=30)) for _ in range(1000)]
plt.hist(samples, bins=30, density=True, edgecolor='black')
plt.title("Central Limit Theorem Applied on Exponential Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
#Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.
sample_means = [np.mean(np.random.normal(loc=0, scale=1, size=30)) for _ in range(1000)]
plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.title("CLT: Sampling Distribution of the Mean")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

```python
# Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).
from scipy.stats import norm

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y)
plt.title("Standard Normal Distribution")
plt.xlabel("Z")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
#Generate random variables and calculate their corresponding probabilities using the binomial distribution.
from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, 11)
pmf = binom.pmf(x, n, p)

for val, prob in zip(x, pmf):
    print(f"P(X={val}) = {prob}")
```

```python
#Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
def calculate_z(x, mean, std):
    return (x - mean) / std

z = calculate_z(75, mean=70, std=10)
print(f"Z-score: {z}")
```

```python
#Implement hypothesis testing using Z-statistics for a sample dataset.
from scipy.stats import norm

sample_mean = 52
population_mean = 50
std_dev = 10
n = 30

z = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 1 - norm.cdf(z)

print(f"Z-statistic: {z}, P-value: {p_value}")
```

```python
#Create a confidence interval for a dataset using Python and interpret the result.
import scipy.stats as stats

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print(f"95% Confidence Interval: {ci}")
```

```python
#Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
import scipy.stats as stats

data = np.random.normal(50, 10, 100)
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print(f"Mean: {mean:.2f}")
print(f"95% Confidence Interval: {ci}")
```

```python
#Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y)
plt.title("PDF of Standard Normal Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
#Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
from scipy.stats import poisson

x = np.arange(0, 20)
cdf = poisson.cdf(x, mu=5)

plt.step(x, cdf, where='post')
plt.title("CDF of Poisson Distribution (Œª=5)")
plt.xlabel("x")
plt.ylabel("CDF")
plt.grid(True)
plt.show()
```

```python
#Simulate a random variable using a continuous uniform distribution and calculate its expected value.
data = np.random.uniform(10, 20, 10000)
expected_value = np.mean(data)
print(f"Expected Value: {expected_value}")

```

```python
# Write a Python program to compare the standard deviations of two datasets and visualize the difference.
data1 = np.random.normal(50, 5, 1000)
data2 = np.random.normal(50, 15, 1000)

std1 = np.std(data1)
std2 = np.std(data2)

print(f"Standard Deviation of Data1: {std1}")
print(f"Standard Deviation of Data2: {std2}")

plt.hist(data1, bins=50, alpha=0.5, label='Std Dev = 5')
plt.hist(data2, bins=50, alpha=0.5, label='Std Dev = 15')
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.show()
```

```python
 #Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
 data = np.random.normal(100, 15, 1000)
range_val = np.max(data) - np.min(data)
q75, q25 = np.percentile(data, [75 ,25])
iqr = q75 - q25

print(f"Range: {range_val}")
print(f"IQR: {iqr}")
```

```python
# Implement Z-score normalization on a dataset and visualize its transformation.
from scipy.stats import zscore

data = np.random.normal(60, 20, 1000)
z_data = zscore(data)

plt.hist(z_data, bins=50, color='orange', edgecolor='black')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
